```python
import FdsPy.qe.myqengine as qe

ts = qe.TimeSeries(start_date = '0',end_date = '0')
univ = qe.IdUniverse(['AAPL-US'])

calc = qe.Calculation(universe = univ,
                      dates = ts,
                    data_dict={'suppliers':'VECTORIZE_MATRIX(FF_COMPANY_RELATIONSHIP(SUPL,ALL,FEID,ALL,ALL))',
                               'partners':'VECTORIZE_MATRIX(FF_COMPANY_RELATIONSHIP(PRTR,ALL,FEID,ALL,ALL))',
                              'customers':'VECTORIZE_MATRIX(FF_COMPANY_RELATIONSHIP(CUST,ALL,FEID,ALL,ALL))'},
                    is_array=True,source = 'FqlExpression')

calc.query()

calc.data
```

```python
import pandas as pd
import numpy as np
df = pd.DataFrame(calc.data['suppliers'].explode()).replace('',np.nan).dropna()
df.insert(1,'value','')
i=7
new_df = pd.DataFrame({'entity_id':df['suppliers'].iloc[::i].values,
                       'entity_name':df['suppliers'].iloc[1::i].values,
                       'rank':df['suppliers'].iloc[2::i].values,
                       'value1':df['suppliers'].iloc[3::i].values,
                       'value2':df['suppliers'].iloc[4::i].values,
                       'source':df['suppliers'].iloc[5::i].values,
                       'value3':df['suppliers'].iloc[6::i].values})
new_df
```
