```python
import pandas as pd
import qe.myqengine as qe
```

```python
univ = qe.ScreenUniverse(universe_expr='((ETP_EXCHANGE(COUNTRY)="US") AND ISAV(ETP_STRUCTURE(ACTIVELY_MANAGE,CODE))=1 AND (ISAV(P_PRICE(0)=1)))=1')
ts = qe.TimeSeries(start_date = '0M')
df = qe.calculate(universe=univ,dates = ts,formulas = ['ETP_CLASS(ASSET,TEXT)',
                                                       'ETP_STRATEGY(STRATEGY,TEXT)',
                                                       'ETP_CLASS(REGION,TEXT)',
                                                       'ETP_STRUCTURE(ACTIVELY_MANAGE,CODE)',
                                                       'ETP_FUND_FLOWS(0,1MON)',
                                                       'ETP_AUM(0,ACT)']).data
df
```
