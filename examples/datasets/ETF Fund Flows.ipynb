```python
import pandas as pd
import qe.myqengine as qe
```

```python
univ = qe.ScreenUniverse(universe_expr='((ETP_EXCHANGE(COUNTRY)="US") AND ISAV(ETP_STRUCTURE(ACTIVELY_MANAGE,CODE))=1 AND (ISAV(P_PRICE(0)=1)))=1')
ts = qe.TimeSeries(start_date = '0M')
data_dict = {"class":'ETP_CLASS(ASSET,TEXT)',
             "strat":'ETP_STRATEGY(STRATEGY,TEXT)',
             "class":'ETP_CLASS(REGION,TEXT)',
             "active_mgt":'ETP_STRUCTURE(ACTIVELY_MANAGE,CODE)',
             "fund_flows_1m":'ETP_FUND_FLOWS(0,1MON)',
             "aum":'ETP_AUM(0,ACT)'}
calc = qe.Calculation(universe=univ,dates = ts,data_dict=data_dict)
calc.query()
calc.data
```
