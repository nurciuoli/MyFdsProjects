```python
import qe.myqengine as qe
```

```python
univ = qe.ScreenUniverse('FG_CONSTITUENTS(SP50,0,CLOSE)=1')
ts = qe.TimeSeries(start_date='0Y-5AY',end_date = '0Y')
df = qe.calculate(universe = univ, dates =ts,formulas = ['CT_TICKER_HIST'])
df
```

```python
df.data
```

```python
unique_tickers = df.data.CT_TICKER_HIST.unique()
```

```python
from misc.utils import get_google_trends
```

```python
import pandas as pd
df_full = pd.DataFrame()
for ticker in unique_tickers:
    df_g = get_google_trends(search_list=[ticker],start_date = '2017-12-29',end_date = '2022-12-30')
    df_full = pd.concat([df_full,df_g.iloc[:,0]],axis=1)

df_full
```

```python
df_full
```

```python
rerun_tickers = set(unique_tickers)- set(df_full.columns)
```

```python
import pandas as pd
df_full2 = pd.DataFrame()
for ticker in rerun_tickers:
    df_g = get_google_trends(search_list=[ticker],start_date = '2017-12-29',end_date = '2022-12-30')
    df_full2 = pd.concat([df_full2,df_g.iloc[:,0]],axis=1)

df_full2
```

```python
df_full=pd.concat([df_full,df_full2],axis=1)
```

```python
rerun_tickers2 = set(rerun_tickers)- set(df_full.columns)
```

```python
import pandas as pd
df_full3 = pd.DataFrame()
for ticker in list(rerun_tickers2)[1:]:
    df_g = get_google_trends(search_list=[ticker],start_date = '2017-12-29',end_date = '2022-12-30')
    df_full3 = pd.concat([df_full3,df_g.iloc[:,0]],axis=1)

df_full3
```

```python
df_full=pd.concat([df_full,df_full3],axis=1)
```
