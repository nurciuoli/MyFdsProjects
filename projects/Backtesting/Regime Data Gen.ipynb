```python
import FdsPy.qe.myqengine as qe
import pandas as pd

univ = qe.ScreenUniverse('FG_CONSTITUENTS(R.3000,0,CLOSE)=1')
ts = qe.TimeSeries(start_date='19970930',end_date = '20161231',frequency='Q',calendar = 'NAY')
data_dict = {"market_val":'ZAV(FREF_MARKET_VALUE_COMPANY(1,USD,1))',
             "3m_fwd_ret":'ZAV(P_PRICE_RETURNS(1,0,+3/0/0,USD))',
            "6m_fwd_ret":'ZAV(P_PRICE_RETURNS(1,0,+6/0/0,USD))',
            "9m_fwd_ret":'ZAV(P_PRICE_RETURNS(1,0,+9/0/0,USD))',
            "12m_fwd_ret":'ZAV(P_PRICE_RETURNS(1,0,+12/0/0,USD))',
             'economy':'AVAIL(QFL_RBICS_CLASS(0,ECON),"OTHER")',
}

calc = qe.Calculation(universe = univ, dates =ts,data_dict=data_dict)
calc.query()
calc.data


df_u = calc.data.copy()

df_u = df_u.set_index(['DATE','UNIVERSE'])

df_u
```
