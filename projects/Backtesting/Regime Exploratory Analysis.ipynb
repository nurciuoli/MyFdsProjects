```python
df_m[['M2','CPI','US_IMP','US_EXP','PMI','GDPN','GDPR']].pct_change().plot(figsize = (20,10))
```

```python
df_flags = df_m[['M2','CPI','US_IMP','US_EXP','PMI','FED_FUND_EFF_RATE','FED_FUND_TAR_RATE']].pct_change().dropna().applymap(lambda x: True if x >0 else False)
```

```python
df_mp = df_m.pct_change().join(df_flags,how='left',rsuffix='_FLAG')
```

```python
df_map = pd.read_csv('../../data/Classifications/rbics_num_reference.csv')
df_map = df_map.rename(columns = {'ID':'economy'})
```

```python
df_map['economy'] = df_map['economy'].astype(str)
```

```python
df_u = df_u.join(df_map.set_index('economy')['NAME'],how='left',on='economy')
```

```python
df= df_u.join(df_mp,how='left',on="DATE")
```

```python
import seaborn as sns
fig,ax = plt.subplots(figsize = (20,15))

sns.violinplot(data = df.reset_index(),x = '3m_fwd_ret',y='NAME',hue = 'FED_FUND_TAR_RATE_FLAG')
```

```python
import seaborn as sns
fig,ax = plt.subplots(figsize = (20,15))

sns.violinplot(data = df.reset_index(),x = '12m_fwd_ret',y='NAME',hue = 'FED_FUND_TAR_RATE_FLAG')
```

```python
df_sector = df.groupby(['DATE','NAME']).agg({'12m_fwd_ret':'mean',
                                             '3m_fwd_ret':'mean'
                                   })

df_sector
```
