```python
import sys
import os
import numpy as np
import pandas as pd
sys.path.append(os.path.dirname(os.path.dirname(os.path.dirname(os.path.abspath('"Top 10 Funds by AUM Positioning - Data Gen.ipynb"')))))
sys.path.append(os.path.dirname(os.path.dirname(os.path.dirname(os.path.abspath('"Top 10 Funds by AUM Positioning - Data Gen.ipynb"')))))
import apis.qe.myqengine as qe
from apis.qe.utils import build_multi_bench_universe_expression,build_multi_conditional_expression,build_multi_bench_formulas
```

```python
indices = ['SP50','R.1000','R.1000G','R.1000V','R.2000','R.2000G','R.2000V','R.3000','R.3000G','R.3000V','CRSPTM']
universe_formula = build_multi_bench_universe_expression(indices)
```

```python
formulas = build_multi_bench_formulas(indices)
```

```python
formulas.append('FREF_MARKET_VALUE_COMPANY(1,USD,1)')
formulas.append('P_PRICE_RETURNS(1,0,+1/0/0,USD)')
formulas.append('P_PRICE_RETURNS(1,0,+3/0/0,USD)')
formulas.append('P_PRICE_RETURNS(1,0,+12/0/0,USD)')
formulas.append('P_PRICE_RETURNS(1,0,+36/0/0,USD)')
formulas
```

```python
time_series = qe.TimeSeries(start_date = '20091231',end_date = '20191231', frequency='M',calendar='NAY')

univ = qe.ScreenUniverse(universe_formula)
req = qe.calculate(universe = univ, dates = time_series,formulas = formulas)

req.data
```

```python
df = req.data.rename(columns = {'DATE':'date','UNIVERSE':'ticker','ZAV(FG_CONSTITUENTS(SP50,0,CLOSE))':'SP50',
                                    'ZAV(FG_CONSTITUENTS(R.1000,0,CLOSE))':'R.1000',
                                    'ZAV(FG_CONSTITUENTS(R.1000G,0,CLOSE))':'R.1000G',
                                    'ZAV(FG_CONSTITUENTS(R.1000V,0,CLOSE))':'R.1000V',
                                    'ZAV(FG_CONSTITUENTS(R.2000,0,CLOSE))':'R.2000',
                                    'ZAV(FG_CONSTITUENTS(R.2000G,0,CLOSE))':'R.2000G',
                                    'ZAV(FG_CONSTITUENTS(R.2000V,0,CLOSE))':'R.2000V',
                                    'ZAV(FG_CONSTITUENTS(R.3000,0,CLOSE))':'R.3000',
                                    'ZAV(FG_CONSTITUENTS(R.3000G,0,CLOSE))':'R.3000G',
                                    'ZAV(FG_CONSTITUENTS(R.3000V,0,CLOSE))':'R.3000V',
                                    'ZAV(FG_CONSTITUENTS(CRSPTM,0,CLOSE))':'CRSPTM',
                                    'FREF_MARKET_VALUE_COMPANY(1,USD,1)':'market_val',
                                    'P_PRICE_RETURNS(1,0,+1/0/0,USD)':'tret_1mf',
                                    'P_PRICE_RETURNS(1,0,+3/0/0,USD)':'tret_3mf',
                                    'P_PRICE_RETURNS(1,0,+12/0/0,USD)':'tret_12mf',
                                    'P_PRICE_RETURNS(1,0,+36/0/0,USD)':'tret_36mf',
                                    

df= df.set_index(['date','ticker'])
df = df[df['market_val']>0]
df
```
